## I Dataset: crx.data

```python
import pandas as pd
import numpy as np

#load Data
data=pd.read_csv('crx.data', header=None)
```

```python
data.head()
```

### In the dataset crx.data, name the columns as col1, col2..col16

```python
col_names =['Col'+str(s) for s in range(1,17)]
data.columns =col_names
data.head()
```

### Print the last 10 rows of data

```python
data.tail(10)
```

### Replace the ‘?’ with Not-a-Number

```python
data=data.replace('?',np.nan)
data.head()
```

### Comment on the datatype of variables

```python
data.info()
```

### The col16 has + and -, replace them ‘P’ and ‘N’ respectively

```python
data['Col16']=data['Col16'].map({'+':'P', '-':'N'})
data.head()
```

```python
data.tail()
```

### Find and display the number of variables of type ‘Object’

```python
obj_columns=[c for c in data.columns if data[c].dtypes=='O']
data[obj_columns].head()
```

```python
print(len(obj_columns))
```

### Display the Month with most of loans issued date

```python
df['date_issued_dt']=pd.to_datetime(df['date_issued'])
df['month']=df['date_issued_dt'].dt.month
df['month'].head()
```

```python
max_val=df['month'].max()
print("Maximum number of loans were issued in the month: ",max_val)
```
